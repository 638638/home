```python
from scipy import stats
# 假设我们关注的是第四年级的学生（假设年级用数字“4”表示）
specific_grade = '大四'

# 筛选出特定年级的数据
df_specific_grade = 认同数据表[认同数据表['年级'] == specific_grade]

# 创建政治面貌和会打多少分的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['中国传统文化'], df_specific_grade['政治面貌'],margins=True,normalize=True)

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

print(chi2_grade,p_grade)
cross_tab_specific_grade
```

分析：
从数据表上可以看出团员占比最高,其次是党员,最后是群众和其他
大四团员对中国传统文化认同度高,认为很有意义,但是有个别认为跟不上潮流
大四党员比较肯定中国传统文化,认同度也很高,无人认为中国传统文化落后
大四群众对中国传统文化认同度也较高,但也存在个别认为中国传统文化落后
